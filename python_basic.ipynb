```python
def find_sum(numbers):
    return sum(numbers)
find_sum([1,2,3])
```

```python
def find_max(numbers):
    maxi = numbers[0]
    for i in numbers():
        if i>numbers:
            maxi = numbers
    return maxi
```

```python
def average_find(marks):
    if len(marks)==0:
        return 0
    return sum(marks)/len(marks)
average_find([1,2,3])
    
```

```python
def cal(numbers):
    even =0
    odd = 0
    for i in numbers:
     if i%2==0:
        even +=1
     else:
        odd+=1
    return (odd,even)

cal([1,2,3])
```
