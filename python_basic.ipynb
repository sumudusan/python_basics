```python
def total(*marks):
    print(marks)

total (5,6,7,8)
```

```python
def tot_marks (*marks):
    total=0
    for mark in marks:
        total +=mark
    print(total)
tot_marks(50,60,70)
```

```python
def tot_marks(**marks):
    print(marks)

tot_marks(math = 20, science=10)
```

List

```python
my_list= ['nimal', 'kamal', 5, 10, [1,2,'sunil']]
print(my_list[0])
print(my_list[-1])
print(my_list[:])
print(my_list[2:])
print(my_list[:-2])
print(my_list[0:2])
print(my_list[1:3])
print(my_list[4][2])

my_list[2]= 'colombo'
print(my_list)

my_list.append('saman')
print(my_list)

my_list.extend('ramal')
print(my_list)

my_list.insert(0,'hello')
print(my_list)

my_list.pop(0)
print(my_list)

my_list.remove('kamal')
print(my_list)

my_list.clear()
print(my_list)

my_list[:]= []
print(my_list)
```

Loops

1.For

```python
x = {'name': 'kamal',
     'age':45,
     'gender':'male'}

for i in x.items():
    print(i)
```

```python
for i in range(5):
    print(i)
```

```python
for i in range(1,6):
    print(i)
```

```python
for x in range(1,6):
    if(x==2):
        continue
    print(x)
```

```python
for x in range(1,6):
    if(x==3):
        break
    print(x)
```

Dictionary

```python
my_dict = {'name':'kamal', 'age':10}
print(my_dict)
print(my_dict['name'])

my_dict['name']= 'sunil'
print(my_dict)

my_dict.update({'gender':'male'})
print(my_dict)

print(my_dict.get('name'))

del my_dict['age']
print(my_dict)

my_dict.pop('name')
print(my_dict)

my_dict.clear()
print(my_dict)
```
