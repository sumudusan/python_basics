```python
def func (name , age=26):
    print(f"my name is {name}. age is {age}")

func('sumudu')
```

```python
def total(*marks):
    print(marks)

total (5,6,7,8)
```

```python
def tot_marks (*marks):
    total=0
    for mark in marks:
        total +=mark
    print(total)
tot_marks(50,60,70)
```

```python
def tot_marks(**marks):
    print(marks)

tot_marks(math = 20, science=10)
```

```python
def student(subject , marks , *friends):
    print('subject =',subject)
    print('marks=',marks)
    print('friends=',friends)
student('maths',45,'nimal, kamal')
```

List

```python
my_list= ['nimal', 'kamal', 5, 10, [1,2,'sunil']]
print(my_list[0])
print(my_list[-1])
print(my_list[:])
print(my_list[2:])
print(my_list[:-2])
print(my_list[0:2])
print(my_list[1:3])
print(my_list[4][2])

my_list[2]= 'colombo'
print(my_list)

my_list.append('saman')
print(my_list)

my_list.extend('ramal')
print(my_list)

my_list.insert(0,'hello')
print(my_list)

my_list.pop(0)
print(my_list)

my_list.remove('kamal')
print(my_list)

my_list.clear()
print(my_list)

my_list[:]= []
print(my_list)
```
